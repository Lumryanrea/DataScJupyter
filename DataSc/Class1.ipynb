# .loc(location) & iloc(index location)

```python
animals = pd.Series(['cat','dog','bird','panda','dragon'], index = [0,3,9,8,3])
animals
```

```python
animals.loc[3]  ##this searches for the location of 3
```

```python
animals.iloc[2] ##this searches for the actual index location
```
