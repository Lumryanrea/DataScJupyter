```python
import pandas as pd
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

## Adding a column(series) to a dataframe

```python
seats = pd.Series([4,4,4,4,4,4])
seats
```

```python
car_sales['Seats'] = seats
car_sales
```

```python
car_sales['Seats'].fillna(4, inplace=True)
```

```python
car_sales
```

## Creating a column from a python list

```python
fuel_economy = [7.5,9.2,5.0,9.6,5.6,6.7,8.2,7.1,9.5,7.5]
car_sales['fuel per 100km'] = fuel_economy
car_sales
#A list must provide a value for all the rows of a column unlike that of series
```

## Calculations

```python
car_sales['Total fuel used'] = car_sales['Odometer (KM)']/100 * car_sales['fuel per 100km']
car_sales
```

## Creating a column from a single value

```python
car_sales['Number of wheels'] = 4
car_sales
```

```python
car_sales['Passed road safety'] =  True
car_sales
```

```python
## checking the datatype
car_sales.dtypes
```

```python
car_sales =  car_sales.drop('Passed road safety', axis=1)
car_sales
```

## shuffling the rows in a dataframe

```python
car_sales
```

```python
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

### this is to reset the dataframe from its shuffled state back to standard by assigning new indexes to the rows


```python
car_sales_shuffled 
```

```python
car_sales_reset = car_sales_shuffled.reset_index()
car_sales_reset
```
